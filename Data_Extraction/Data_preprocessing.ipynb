```python
from pandas import read_excel
import numpy as np
import matplotlib.pyplot as plt


my_sheet = 'Data_preprocessing'
file_name = 'SemanticScholarJournalData.xlsx'
paper_df = read_excel(file_name, sheet_name = my_sheet,parse_cols = "A,E,H,N:O")

counts = paper_df['Domain'].value_counts().to_dict()
print(counts)
domain_plot = plt.bar(range(len(counts)), list(counts.values()), align='center',animated = True)
domain_plot= plt.xticks(range(len(counts)), list(counts.keys()))
domain_plot = plt.rcParams["figure.figsize"] = [12,5]
plt.xlabel('Domain of papers', fontsize=18)
plt.ylabel('Number of papers', fontsize=16)
plt.show()
```

```python
from pandas import read_excel
import numpy as np
import matplotlib.pyplot as plt


my_sheet = 'Data_preprocessing'
file_name = 'SemanticScholarJournalData.xlsx'
paper_df = read_excel(file_name, sheet_name = my_sheet,parse_cols = "N,H")
sumCitation = paper_df.groupby(['Technology']).sum().to_dict()
sumCitation_dict = sumCitation['Citations']
print (sumCitation_dict)

Citations_plot = plt.bar(range(len(sumCitation_dict)), list(sumCitation_dict.values()), align='center',animated = True)
Citations_plot= plt.xticks(range(len(sumCitation_dict)), list(sumCitation_dict.keys()))
Citations_plot = plt.rcParams["figure.figsize"] = [12,5]
plt.xlabel('Technology', fontsize=18)
plt.ylabel('Number of Citations', fontsize=16)
plt.show()
```
